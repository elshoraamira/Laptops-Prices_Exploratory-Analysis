## Investigation Overview

In this investigation, I wanted to look at the components of laptops that could be used to predict their prices. The main focus was on the main operation components: Processor Type (CPU), RAM Size (in GB) and GPU Brand.

## Dataset Overview

The data consisted of prices and attributes of approximately 1300 laptops. The attributes included my predictors, as well as additional features such as type, size, and weight. Sixty-four data points were removed from the analysis due to inconsistencies or duplicate information.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
laptops = pd.read_csv('laptops.csv')
```

```python
#dropping "Unnamed: 0" column
laptops.drop('Unnamed: 0', axis=1, inplace=True)
#dropping duplicates 
laptops = laptops.drop_duplicates()
```

```python
#remove 'kg' by replacing it with empty then change its data type to int 
laptops['Weight'] = [sub.replace("kg", "").strip() for sub in laptops['Weight']]
laptops['Weight'] = pd.to_numeric(laptops['Weight'])
#extract GPU company producer name from "Gpu" column and add it to new column "Gpu_Brand"
laptops['Gpu_Brand'] = laptops['Gpu'].str.split(' ').str[0]
#extract CPU type from "Cpu" column and add it to new column "Cpu_Type"
laptops['Cpu_Type'] = (laptops['Cpu'].str.split(' ').str[0] + " " + laptops['Cpu'].str.split(' ').str[1])
```

```python
#Dropping outliers
#laptops_copy
laptops_copy = laptops
li_drop = ['AMD E-Series', 'AMD A12-Series', 'AMD A10-Series', 'AMD Ryzen', 'Intel Xeon', 'AMD A8-Series', 'AMD FX', 'Samsung Cortex', 'AMD A4-Series']
laptops_copy = laptops_copy[~laptops_copy['Cpu_Type'].isin(li_drop)]
```

## Distribution of Laptops Prices

Laptops prices in the dataset take on a very large range of values, from about 170 at the lowest to about 6100 at the highest. Plotted on a logarithmic scale, the distribution of diamond prices takes on a unimodal shape.

```python
bins = np.logspace(np.log10(laptops_copy['Price_euros'].min()),np.log10(laptops_copy['Price_euros'].max()), 17)
plt.figure(figsize=[8, 5])
plt.hist(data = laptops, x = 'Price_euros', bins = bins)
plt.xscale('log')
# plt.xticks(list(bins), [int(num) for num in list(bins)])
plt.xlim([100, 7000])
plt.xticks([200, 500, 1e3, 2e3, 4e3, 6e3], [200, 500, '1k', '2k','4k', '6k'], fontsize=11)
plt.yticks(fontsize=11)
plt.xlabel('Price (â‚¬)', fontsize=12)
plt.title('Distribution of Laptop Prices', fontsize=15, fontweight="bold")
plt.show()
```

## Price VS. CPU

Most expensive laptops are those with Inter Core processor <br>
Laptops with other processor types are almost equal in price <br>

```python
#setting the color
default_color = sb.color_palette()[0]

plt.figure(figsize = [10, 6])
cpu_order = laptops_copy['Cpu_Type'].value_counts().index.tolist()
sb.swarmplot(data = laptops_copy, x = 'Cpu_Type', y = 'Price_euros', color = default_color, order = cpu_order);

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.title('Price VS. CPU', fontsize=15, fontweight="bold")
plt.xlabel('CPU Type', fontsize=12)
plt.ylabel('Price (â‚¬)', fontsize=12);
```

## Price VS. RAM

Most expensive laptops are those with 64GB RAM <br>
Laptops with 8GB are the most common as well as with average price <br>
Laptops with 2GB are the cheapest of them all <br>

```python
plt.figure(figsize = [10, 6])
ram_orders = ['2GB', '4GB', '6GB', '8GB', '12GB', '16GB', '24GB', '32GB', '64GB']
sb.swarmplot(data = laptops_copy, x = 'Ram', y = 'Price_euros', color = default_color, order=ram_orders);

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.title('Price VS. RAM', fontsize=15, fontweight="bold")
plt.xlabel('RAM Size', fontsize=12)
plt.ylabel('Price (â‚¬)', fontsize=12);
```

## Price VS. CPU, RAM, GPU & Type

CPU, GPU & Type record a price range of hundreds to 2000, On the other hand the RAM curve is rising dramatically to price ranges up to 4000, For that we can expect that RAM is a very influential factor in setting price for a laptop.
(The larger the RAM size the more expensive the laptop is)

```python
replacements = {'Cpu_Type': 'CPU Type', 'Ram': 'RAM Size',
                'Gpu_Brand': 'GPU Brand', 'TypeName': 'Laptop Type'}
g = sb.PairGrid(laptops_copy, y_vars="Price_euros",
                 x_vars=["Cpu_Type", "Ram", "Gpu_Brand", "TypeName"], height=4, aspect=1)
g.map(sb.pointplot, color=default_color);
for ax in g.axes.flat: 
    ax.set_xticklabels(ax.get_xticklabels(), rotation=25, fontsize=11)
    
for i in range(1):
    g.axes[0][0].set_ylabel("Price (â‚¬)", fontsize=13)
    for j in range(4):
        xlabel = g.axes[i][j].get_xlabel()
        ylabel = g.axes[i][j].get_ylabel()
        if xlabel in replacements.keys():
            g.axes[i][j].set_xlabel(replacements[xlabel], fontsize=13)
            
plt.subplots_adjust(top=0.85)          
g.fig.suptitle('Price VS. CPU Type - RAM Size - GPU Brand - Laptop Type', fontsize=16, fontweight="bold");
```

## Price by RAM & CPU

The most expensive Laptop is one with CPU of Type Intel Core and 64GB RAM <br>
The cheapest laptops are those with CPU of the other types with RAM size of 2 - 8 GB

```python
g = sb.catplot(x="Ram", y="Price_euros",
                hue="Cpu_Type",
                data=laptops_copy, kind="point",
               order = ram_orders,
                dodge=True,
               linestyles = '',
                height=4, aspect=1.5, legend = False);
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.title('Price by RAM & CPU', fontsize=15, fontweight="bold")
plt.xlabel('RAM Size', fontsize=12)
plt.ylabel('Price (â‚¬)', fontsize=12);
plt.legend(title="CPU Type", fontsize=9);
```

## Price by RAM & CPU for GPU Brands

Nivida GPU comes with Intel processor almost in all cases <br>
The most expensive Laptop is one with CPU of Type Inel Core, 64GB RAM and GPU Brand Nividia <br>
The cheapest laptops are those with CPU of the other types with RAM size of 2 - 8 GB and GPU Brand AMD

```python
g = sb.catplot(x="Ram", y="Price_euros",
                hue="Cpu_Type", col="Gpu_Brand",
                data=laptops_copy, kind="point",
               order = ram_orders,
                dodge=True,
               linestyles = '',
                height=4, aspect=1.1);

for i in range(3):
    g.axes[0][i].set_xlabel("RAM Size", fontsize=13)
    g.axes[0][0].set_ylabel('Price (â‚¬)', fontsize=13)
# plt.xticks(fontsize=11)
# plt.yticks(fontsize=11)
# # plt.title('Price by RAM & CPU', fontsize=15, fontweight="bold")
# plt.xlabel('RAM Size', fontsize=12)
# plt.ylabel('Price (â‚¬)', fontsize=12);

g.set_titles(col_template = '{col_name}')
g._legend.set_title("CPU Type")
plt.subplots_adjust(top=0.85)          
g.fig.suptitle('Price by RAM & CPU for GPU Brands', fontsize=16, fontweight="bold");
```
